```python
import pandas as pd
import  matplotlib.pyplot as plt
import numpy as np
```

```python
difference_scores = pd.read_csv('../data/processed/difference_volt_per_neighboorhoud.csv')
election_2021 = pd.read_csv('../data/raw/uitslag_2021_landelijk.csv', index_col = 0)
election_2022 = pd.read_csv('../data/raw/uitslag_2022_gemeenteraad.csv', index_col = 0)
election_2021.neighborhood.value_counts().keys() == election_2022.neighborhood.value_counts().keys() #check if neighborhoods are the same
```

```python
fig = plt.figure()
fig = fig.add_axes([0,0,1,1])
votes_per_election = [sum(election_2021.votes), sum(election_2022.votes)]
plt.bar(['national elections 2021', 'council elections 2022'],votes_per_election)
fig.set_title('Turnout elections')
plt.savefig('../data/graphs/turnout_elections.png')
plt.show
```

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
# set height of bar
votes_per_election = [sum(election_2021.votes), sum(election_2022.votes)]
eligible_voters_per_election = [356_481, 428_137]
 
# Set position of bar on X axis
br1 = np.arange(len(votes_per_election))
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, votes_per_election, color ='#5F4B8BFF', width = barWidth,
         label ='turnout')
plt.bar(br2, eligible_voters_per_election, color ='#E69A8DFF', width = barWidth,
        edgecolor ='grey', label ='eligible voters')

 
# Adding Xticks
plt.xlabel('election', fontweight ='bold', fontsize = 15)
plt.ylabel('amount of people', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(votes_per_election))],
       ['national elections 2021', 'council elections 2022'])
 
plt.legend(fontsize = 14)
plt.savefig('../data/graphs/turnout_vs_eligble.png')
plt.show()
```
