```python
from selenium import webdriver 
import requests
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import pandas as pd
import time
from collections import ChainMap
```

```python
driver = webdriver.Chrome()
driver.get('https://tweedekamer2021.stemwijzer.nl/#/')
driver.find_element(by=By.CSS_SELECTOR, value='body > div > div > div > div.start__modal-left > div.start__modal > div.start__modal-footer > button').click()
#driver.find_element_by_css_selector('body > div > div > div.statement__inner > div > button.statement__tab-button.statement__tab-button--parties').click()
votes_per_party = []


for i in range(30):
    try:
        if len(votes_per_party) == 30:#break if amount of statements is the same as the amount of statements on the stemwijzer site
                    break
        time.sleep(1)
        question = driver.find_element(by = By.CSS_SELECTOR, value = 'body > div > div > div.statement__inner > h1').text #select the statement element
        time.sleep(1)
        driver.find_element(by = By.CSS_SELECTOR, value = 'body > div > div > div.statement__inner > div > button.statement__tab-button.statement__tab-button--parties').click()#open the tab that contains the opinion of the parties
        time.sleep(1)
        agrees = driver.find_element(by = By.CSS_SELECTOR, value = 'body > div > div > div.statement__tab.statement__tab--parties > div:nth-child(1)').text.split('\n')[2:] #all parties that agree with the statement
        time.sleep(1)
        neutral =  driver.find_element(by = By.CSS_SELECTOR, value = 'body > div > div > div.statement__tab.statement__tab--parties > div:nth-child(2)').text.split('\n')[2:]#all parties that are neutral on the statement
        time.sleep(1)
        disagrees =  driver.find_element(by = By.CSS_SELECTOR, value = 'body > div > div > div.statement__tab.statement__tab--parties > div:nth-child(3)').text.split('\n')[2:]#all parties that disagree with the statement
        time.sleep(1)
        agrees_dict = {party : 1 for party in agrees}#score of 1 if the party agrees with the statement
        disagrees_dict = {party : -1 for party in disagrees}#score of 0 if the party agrees with the statement
        neutral_dict = {party : 0 for party in neutral}#score of -1 if the party agrees with the statement
        question_dict = {'question': question}
        full_dict = dict(question_dict, **agrees_dict, **disagrees_dict , **neutral_dict) #merge the dicts
        if len(votes_per_party) > 0: #otherwise and index-error occurs.
            if votes_per_party[len(votes_per_party) - 1]['question'] == question:#compare the block to previous statement. Sometimes, the scraper fails to go to the next page. this avoids saving duplicate questions
                print('failed to go to next page')
                raise ValueError
        votes_per_party.append(full_dict)#add dict to list to turn into pandas df later

        #sometimes, a error occurs when trying to find click the element that allows the scraper to go to the next question. two different elements are tried in the lines below. if they fail, the script retries for 10 rounds
        for i in range(10):
            try:
                driver.find_element(by = By.CSS_SELECTOR, value ='body > div > div > div.statement__buttons > div.statement__buttons-mobile > div > div:nth-child(2) > button').click()
                break
            except:
                try:
                    driver.find_element(by = By.CSS_SELECTOR, value = 'body > div > div > div.statement__buttons.statement__buttons--floating > div.statement__buttons-mobile > div > div:nth-child(2) > button').click()
                    break
                except:
                     pass
```

```python
pdf_votes_per_party_stemwijzer_2021 = pd.DataFrame(votes_per_party)#turn list of dicts from scraper into a pandas df
```
